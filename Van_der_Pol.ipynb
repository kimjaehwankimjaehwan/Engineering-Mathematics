```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Van der Pol  
def van_der_pol(X, t, mu):
    x1, x2 = X  # x1: , x2: 
    dx1dt = x2
    dx2dt = mu * (1 - x1**2) * x2 - x1
    return [dx1dt, dx2dt]

#     
mu = 1.0  # Van der Pol  ( )
initial_conditions = [2.0, 0.0]  #  [x1(0), x2(0)]

#   
t = np.linspace(0, 20, 1000)  # 0 20 1000 

# odeint   
solution = odeint(van_der_pol, initial_conditions, t, args=(mu,))

#  
x1 = solution[:, 0]  #   (x1)
x2 = solution[:, 1]  #   (x2)


plt.figure(figsize=(10, 5))
plt.plot(t, x1, label='Position (x1)')
plt.plot(t, x2, label='Velocity (x2)', linestyle='dashed')
plt.title('Van der Pol Oscillator (mu={})'.format(mu))
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```
