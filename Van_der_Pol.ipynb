```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Van der Pol 방정식 정의
def van_der_pol(X, t, mu):
    x1, x2 = X  # x1: 위치, x2: 속도
    dx1dt = x2
    dx2dt = mu * (1 - x1**2) * x2 - x1
    return [dx1dt, dx2dt]

# 초기 조건 및 파라미터 설정
mu = 1.0  # Van der Pol 파라미터 (비선형성 제어)
initial_conditions = [2.0, 0.0]  # 초기값 [x1(0), x2(0)]

# 시간 범위 설정
t = np.linspace(0, 20, 1000)  # 0에서 20초까지 1000개의 시간점

# odeint를 이용한 방정식 풀이
solution = odeint(van_der_pol, initial_conditions, t, args=(mu,))

# 결과 출력
x1 = solution[:, 0]  # 위치 값 (x1)
x2 = solution[:, 1]  # 속도 값 (x2)


plt.figure(figsize=(10, 5))
plt.plot(t, x1, label='Position (x1)')
plt.plot(t, x2, label='Velocity (x2)', linestyle='dashed')
plt.title('Van der Pol Oscillator (mu={})'.format(mu))
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```
